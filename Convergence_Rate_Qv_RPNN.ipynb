```julia
using LinearAlgebra
using PyPlot
using Random
using Statistics
using Printf
using JLD2, FileIO
using ProgressMeter
using Quaternions
include("HyperRPNNs.jl")
include("HyperECNNs.jl")
rng = MersenneTwister(0);
```

```julia
MarkerList = ["o","s","v","^",">","<"];
LineStyleList = ["solid","dashed","solid","dashed","solid","dashed"];
```

```julia
function imnoise(img,noise_std = 0.1)
    return clamp.(img + noise_std*randn(rng, Float64, size(img)),-1,1)
end
```

```julia
function plotseq(Sequence,ind,N)
    ### Convergence Ratio ###
    
    num_elements = length(Sequence)
    
    #print(size(Sequence[1]))
    
    Sequencia_c = zeros(N,4,num_elements)
    
    for i=1:num_elements
        Sequencia_c[:,:,i] = Sequence[i][:,:]
    end
    
    Convergence_rate = zeros(num_elements-1,1)
    
    for j=1:num_elements-1
       Convergence_rate[j] = norm(Sequencia_c[:,:,j+1]-Sequencia_c[:,:,j])/norm(Sequencia_c[:,:,j+1])
    end
    
    c1 = 0
    k = ind
    if k>6
        c1 = c1+1
        k = c1
    end
    
    c2 = 0
    v = ind
    if v>6
        c2 = c2+1
        v = c2
    end 
    
    return k,v,Convergence_rate,num_elements
end
```

```julia
N = 100
P = 160
K = 16
alpha = 1
beta = -1

MultiStates = exp.(2*pi*collect(0:K-1)*im/K)
U=zeros(N,4,P)


Uz1 = rand(rng,MultiStates,(N,P))
Uz2 = rand(rng,MultiStates,(N,P))
U[:,1,:] = real.(Uz1)
U[:,2,:] = imag.(Uz1)
U[:,3,:] = real.(Uz2)
U[:,4,:] = imag.(Uz2)

img = U[:,:,1]
xcopy = copy(img)

for i=2:2
    
    noise = 0.33*i
    
    x = imnoise(img,noise)
    print(norm(xcopy-x))
    
    y, Sequence = HyperRPNNs.Sync(HyperRPNNs.LambdaInner,[1,1,1,1],HyperRPNNs.twincsign,K,U,x,alpha,beta,5)
    y, Sequence2 = HyperECNNs.Sync(HyperECNNs.LambdaInner,[1,1,1,1],HyperECNNs.twincsign,K,U,x,alpha/(2*N),beta,5)
    
    #print(length(Sequence))
    #print(size(Sequence[1]))
    
    k,v,Convergence_rate,num_elements = plotseq(Sequence,i,N)
    k2,v2,Convergence_rate2,num_elements2 = plotseq(Sequence2,i,N)
    
    plot(collect(1:num_elements-1),Convergence_rate, linestyle = LineStyleList[1], marker = MarkerList[k], label = "EPNN")
    plot(collect(1:num_elements2-1),Convergence_rate2, linestyle = LineStyleList[2], marker = MarkerList[k],  mfc="none",  label = "ECNN")
end
xlabel("t")
ylabel("Convergence Ratio")
title("Quaternion-Valued MultiState Model")
legend()
grid()
savefig("Figures/Conv_rate_QvComplexRPNN.eps")
savefig("Figures/Conv_rate_QvComplexRPNN.pdf")
```
